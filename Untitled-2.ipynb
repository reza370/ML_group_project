```python
import pandas as pd
import numpy as np

# Create a DataFrame from the provided data
data = {
    'Egg price': [1, 1, 2, 2, 3, 3],
    'Gold price': [1, 2, 2, 3, 3, 4],
    'Oil price': [2, 1, 3, 2, 4, 3]
}

df = pd.DataFrame(data)

# Define the initial centroids
centroids = np.array([[2, 2, 2], [1, 2, 3], [3,5, 3]])

# Perform two iterations of K-means clustering
for _ in range(10):
    # Calculate the distances between each data point and the current centroids
    distances = np.zeros((len(df), len(centroids)))
    for i, row in df.iterrows():
        for j, centroid in enumerate(centroids):
            distances[i, j] = np.linalg.norm(row - centroid)

    # Assign each data point to the nearest centroid
    clusters = np.argmin(distances, axis=1)

    # Calculate the new centroids
    new_centroids = np.zeros_like(centroids)
    for cluster in range(len(centroids)):
        cluster_data = df[clusters == cluster]
        new_centroids[cluster] = cluster_data.mean().to_numpy()

    # Update the centroids
    centroids = new_centroids

# Print the final centroids
print("Final centroids after two iterations:")
print(centroids)
```

```python
from sklearn.metrics import pairwise_distances_argmin
import numpy as np
import matplotlib.pyplot as plt
X = np.array([[2, 3], [1, 2], [6, 7]]) #close to 1, 0, 2 centers
centers = np.array([[1, 1], [2, 2], [3, 3]])

pairwise_distances_argmin(X, centers)
#Implement K-means from scratch
from sklearn.datasets import make_blobs
from sklearn.metrics import pairwise_distances_argmin
from time import time

X, y_true = make_blobs(n_samples=1500, centers=4,
                       cluster_std=0.60, random_state=0)

def kmeans(X, n_clusters):
    m, n = X.shape

    #1. randomly choose n clusters from X
    #you can also randomly generate any two points
    rng = np.random.RandomState(42)
    i = rng.permutation(m)[:n_clusters]
    centers = X[i]

    iteration = 0

    while True:
        #2. assign lables based on closest center
        #return the index of centers having smallest
        #distance with X
        labels = pairwise_distances_argmin(X, centers)

        #3. find new centers
        new_centers = []
        for i in range(n_clusters):
            new_centers.append(X[labels == i].mean(axis=0))

        #convert list to np.array; you can actually combine #3
        #with np.array in one sentence 
        new_centers = np.array(new_centers)
        
        #compute total within-variation score
        total_with_variation_score = 0
        labels = pairwise_distances_argmin(X, centers) #<---Note I use X here.  Why?
        for i in range(n_clusters):
            cluster_mean = X[labels==i].mean(axis=0)
            total_with_variation_score += ((X[labels==i] - cluster_mean)** 2).sum()
            
        print("Total with variation score: ", total_with_variation_score)
        
        #plotting purpose
        #plot every 5th iteration to save space
        #remove this if, if you want to see each snapshot
        if (iteration % 5 == 0):
            pred = pairwise_distances_argmin(X, new_centers)
            plt.figure(figsize=(5, 2))
            plt.title(f"Iteration: {iteration}")
            plt.scatter(X[:, 0], X[:, 1], c=pred)
            plt.scatter(new_centers[:, 0], new_centers[:, 1], s=100, c="black", alpha=0.6)

        #4 stopping criteria - if centers do not 
        #change anymore, we stop!
        if(np.allclose(centers, new_centers)):
            break
        else:
            centers = new_centers
            iteration+=1
    
    print(f"Done in {iteration} iterations")
    return centers
        
def predict(X, centers):
    return pairwise_distances_argmin(X, centers)

start = time()
preds = predict(X, kmeans(X, n_clusters=4))
print(f"Fit and predict time: {time() - start}")
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=preds, s=50)
plt.title("Final result")
```

```python
Wh = torch.randn(3,3, requires_grad = True)
Wx = torch.randn(3,2, requires_grad = True)
h = torch.randn(1,3)
x = torch.randn(1,2)

wh2h = torch.mm(Wh,h.t())
wx2x = torch.mm(Wx,x.t())
print(wh2h)
print(wx2x)

next_h = wh2h + wx2x + 5
next_h = next_h.tanh()
print(next_h)

loss = next_h.sum()
print(loss)

# backprop
loss.backward()

# stores gradients when requires_grad = True
print(Wx.grad)
print(Wh.grad)
# does not store gradients when requires_grad = False
print(x.grad)
print(h.grad)
```
